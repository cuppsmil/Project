### **Посчитайте Retention 15 дня (в процентах) для пользователей, зарегистрированных в январе**

Cохраните результат в переменную `retention_15_day`

**Примечание:** результат округлите до 5 знаков после запятой

```python
import pandas as pd
#Читаем csv
entries = pd.read_csv('entries.csv', sep=';')
registrations = pd.read_csv('registrations.csv', sep=';')
#Переводим колонки дат в тип datetime
registrations['registration_date'] = pd.to_datetime(registrations['registration_date'])
entries['entry_date'] = pd.to_datetime(entries['entry_date'])
#Фильтруем даты регистрации по январю
regs_in_january = registrations[registrations['registration_date'].dt.month == 1]
#Джоиним таблицы по user_id, inner join
students_with_entries = entries.merge(regs_in_january, on='user_id', how='inner')
#Считаем разницы между датой регистрации и датами заходов
students_with_entries['diff'] = (students_with_entries['entry_date'] - students_with_entries['registration_date']).dt.days
#Ищем всех пользователей, которые зашли в 15 день
students_15_day = students_with_entries[students_with_entries['diff'] == 15]
#Ищем всех пользователей, которые зашли в 0 день (зарегались)
students_0_day = students_with_entries[students_with_entries['diff'] == 0]
#Считаем уникальные ID
cnt_15 = students_15_day['user_id'].nunique()
cnt_0 = students_0_day['user_id'].nunique()
#Retention 15 дня
retention_15_day = round(cnt_15 / cnt_0 * 100, 5)



```

```python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
# Открываем файл с правильными ответами
with open('metrics.txt', 'r') as f:
    answers = f.read().split('\n')

correct_answer = float(answers[0])

try:
    assert retention_15_day == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```
